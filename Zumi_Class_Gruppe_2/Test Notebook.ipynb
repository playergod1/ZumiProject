# Zumi random walk and following a graph route

```python
from ZumiRPC import Zumi
from Route_calculation import Navigator
import networkx as nx
```

Initialize Zumi and read graphml into graph. Then get instance of Navigator using zumi and graph

```python
zumi = Zumi(direction =-100,port = 9002) #current direction must given, 0 = right, 90 = down, 180 = left, -90 = up
G = nx.read_graphml('third_try.graphml')
```

```python
navi = Navigator(zumi,G)
```

## To drive towards point, simply call drive_towards(x,y). The class will take care of all driving commandos and calculations. 

```python
navi.drive_towards(0,0)
```
